```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Ecommerce Customers")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# EDA
sns.jointplot(x="Time on Website", y="Yearly Amount Spent", data=data, alpha=0.5)
```

```python
sns.jointplot(x="Time on App", y="Yearly Amount Spent", data=data, alpha=0.5)
```

```python
sns.pairplot(data, kind="scatter", plot_kws={"alpha" : 0.4})
# to check the correlation between variables
```

```python
sns.lmplot(x="Length of Membership",
           y= "Yearly Amount Spent",
           data=data,
            scatter_kws={"alpha":0.3})
```

```python
from sklearn.model_selection import train_test_split
```

```python
#X is features while y is target variable
X=data[["Avg. Session Length", "Time on App", "Time on Website", "Length of Membership"]]
y=data["Yearly Amount Spent"]
```

```python
X_train, X_test, y_train, y_test=train_test_split(X ,y, test_size=0.3, random_state=42)
```

```python
X_test
```

```python
# training the model
from sklearn.linear_model import LinearRegression
ln=LinearRegression()
```

```python
ln.fit(X_train, y_train)
```

```python
cdf=pd.DataFrame(ln.coef_, X.columns, columns={"Coef"})
print(cdf)
```

```python
# Predictions
predictions=ln.predict(X_test)
predictions
```

```python
sns.scatterplot(predictions, y_test)
plt.xlabel("Predictions")
plt.title("Evaluation of LRM")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

print("Mean Absolute Error:", mean_absolute_error(y_test, predictions))
print("Mean Squared Error:", mean_squared_error(y_test, predictions))
print("RMSE:", math.sqrt(mean_squared_error(y_test, predictions)))


```

```python
#Residuals
residuals=y_test - predictions
sns.displot(residuals, bins=30, kde=True)
```
